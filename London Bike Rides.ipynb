Specify a dictionary for the column names I want to use

```python
new_cols_dict ={
    'timestamp':'time',
'cnt':'count',
't1':'temp_real_C',
't2':'temp_feels_like_C',
'hum':'humidity_percent',
'wind_speed':'wind_speed_kph',
'weather_code':'weather',
'is_holiday':'is_holiday',
'season':'season'
}
```

Rename the columns to the newly specified column names

```python
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

Let's change the humidity values to percentages (values between 0 and 1)

```python
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
bikes
```

Let's create a couple more dictionaries
The first one will be for seasons, so that we can map the integers 0-3 to the actual written values
The second one will be for weather, so that we can do the same.

```python
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter',
}

weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rainy',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall',
}
```

```python
bikes
```

```python
#Change the seasons column data type to string
bikes.season = bikes.season.astype('str')
#Map the values 0-3 to the actual written seasons
bikes.season = bikes.season.map(season_dict)
```

```python
#Change the weather column data type to string
bikes.weather = bikes.weather.astype('str')
#Map the values 0-3 to the actual written seasons
bikes.weather = bikes.weather.map(weather_dict)
```
